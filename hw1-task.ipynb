```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10 000
# принимает значение B, если значение LIMIT_BAL <=100 000 и >10 000
# принимает значение C, если значение LIMIT_BAL <=200 000 и >100 000
# принимает значение D, если значение LIMIT_BAL <=400 000 и >200 000
# принимает значение E, если значение LIMIT_BAL <=700 000 и >400 000
# принимает значение F, если значение LIMIT_BAL >700 000

pd.DataFrame([df['LIMIT_BAL'], df['LIMIT_BAL'].map(lambda x: x <= 20000, 'A' )]).T
def make_new_column(x):
    y=''
    if x<=10000:
        y = 'A'
    elif x<=100000:
        y = 'B'
    elif x<=200000:
        y = 'C'
    elif x<=400000:
        y = 'D'
    elif x<=700000:
        y = 'E'
    else:
        y = 'F'
    return y
```

```python
df['LIMIT_BAL_litera'] = df['LIMIT_BAL'].copy().apply(make_new_column)
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# расчёт матрицы корреляций Пирсона
corr = df.corr(method = 'pearson')


mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# расчёт матрицы корреляций Кендалла
corr = df.corr(method = 'kendall')

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# расчёт матрицы корреляций Спирмена
corr = df.corr(method = 'spearman')

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
